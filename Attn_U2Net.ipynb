```python
import os
import glob

folder_path = "/content/drive/MyDrive/Attention-U2Net-main/Attention U2Net/train_data/DUTS_full/DUTS_testimg"

# Use glob to find all JPEG and PNG files in the folder
jpeg_files = glob.glob(os.path.join(folder_path, "*.jpg"))
# png_files = glob.glob(os.path.join(folder_path, "*.png"))

# Count the total number of JPEG and PNG files
num_images = len(jpeg_files)

print("Number of images:", num_images)
```

```python
import os
import random
import shutil

def select_random_images_and_labels(src_img_folder, src_label_folder, dst_img_folder, dst_label_folder, num_images, start_index=0):
    img_files = os.listdir(src_img_folder)
    random.shuffle(img_files)
    
    if not os.path.exists(dst_img_folder):
        os.makedirs(dst_img_folder)
    
    if not os.path.exists(dst_label_folder):
        os.makedirs(dst_label_folder)
    
    for i in range(start_index, start_index + num_images):
        img_file = img_files[i]
        img_basename, img_ext = os.path.splitext(img_file)
        label_file = img_basename + ".png"  # Assuming the labels have .png extension

        shutil.copy(os.path.join(src_img_folder, img_file), os.path.join(dst_img_folder, img_file))
        shutil.copy(os.path.join(src_label_folder, label_file), os.path.join(dst_label_folder, label_file))

src_img_folder = "DUTS-TR/DUTS-TR-Image"
src_label_folder = "DUTS-TR/DUTS-TR-Mask"
dst_img_folder = "DUTS_imgs"
dst_label_folder = "DUTS_labels"
dst_test_img_folder = "DUTS_testimg"
dst_test_label_folder = "DUTS_testlabel"
num_images = 5000
num_test_images = 1000

# Select 5000 images for DUTS_imgs and DUTS_labels
select_random_images_and_labels(src_img_folder, src_label_folder, dst_img_folder, dst_label_folder, num_images)

# Select additional 1000 images for DUTS_testimg and DUTS_testlabel, different from the earlier 5000 images
select_random_images_and_labels(src_img_folder, src_label_folder, dst_test_img_folder, dst_test_label_folder, num_test_images, start_index=num_images)
```
